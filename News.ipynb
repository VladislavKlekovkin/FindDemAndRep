```python
"""
Определяет относится ли новость к какой-то из партий
Программа нашла только одно совпадение оно лежит в файле rep.txt
"""

old_url = ''
import time
import re
import requests
import bs4

while True:
    
    
    # ищем новые статьи
    fnew_url = "https://edition.cnn.com/specials/last-50-stories"
    first_old_url = "https://edition.cnn.com/2019/03/21/us/indiana-active-shooter-pellets/index.html"
    if old_url == '':
        old_url = first_old_url
    r = requests.get(fnew_url)
    
    list_of_new_url = []
    soup = bs4.BeautifulSoup(r.text, "html.parser")

    for tag in soup.find_all():
        if tag.name == 'a':
            try:
                t =  tag['href']
            except KeyError:
                t = False
        
            if t != False and '2019' in t and 'videos' not in t:
                result = 'https://edition.cnn.com' + t
                if result != old_url:
                    list_of_new_url.append(result)
                    print(result)
                else:
                    break
    
    
    # Анализируем полученные ссылки
    
    for i in range(len(list_of_new_url)):
        
        r = requests.get(list_of_new_url[i])
        soup = bs4.BeautifulSoup(r.text, "html.parser")
        
        # берем название статьи
        headline = re.search('<h1 class="pg-headline">(.*)</h1>', r.text)
        #print(result.group(1))
        
        # берем авторов статьи
        for tag in soup.find_all():
            if tag.name == 'span':
                try:
                    c =  tag['class']
                   
                except KeyError:
                    c = False
                    
                if c == ["metadata__byline__author"]:
                    if 'a href' in tag.text:
                        a = re.search('>(.*)</a>', tag.text)
                        
                        authors = a.group(1)
                    else:
                        
                        #a = re.search('<span class="metadata__byline__author">(.*)CNN</span>', tag.text)
                        authors = tag.text[:len(tag.text)-5]
                        #print(authors)
                        
                   
                    break
                
                
        #print(autors.group(1)[:len(autors.group(1))-2])
        
        # ищем ключевые слова

        for tag in soup.find_all():
            #print(tag.attrs)
            try:
                t = (tag['class'] == ["zn-body__paragraph"])
            except KeyError:
                t = False

            if tag.name == 'div' and t:
                
                #
                #print('Я тут!')
                
                if ('democrat' in tag.text or 'Dem\'s' in tag.text) and ('republican' in tag.text or 'GOP' in tag.text):
                    
                    with open('C:/Users/user/both.txt', 'a') as file:
                        file.write(str(headline.group(1)) + '\n' + str(authors) + '\n' + 
                                   str(list_of_new_url[i]) + '\n\n')
                    
                    print(str(headline.group(1)) + '\n' + str(authors) + '\n' + 
                        str(list_of_new_url[i]) + '\n\n')
 
                    break
                else:
                    if ('democrat'  in tag.text or 'Dem\'s'  in tag.text):
                        
                        with open('C:/Users/user/dem.txt', 'a') as file:
                            file.write(str(headline.group(1)) + '\n' + str(authors) + '\n' + 
                                    str(list_of_new_url[i]) + '\n\n')
                        print(str(headline.group(1)) + '\n' + str(authors) + '\n' + 
                            str(list_of_new_url[i]) + '\n\n')
                        
                        break
                    
                    if ('republican'  in tag.text or 'GOP' in tag.text):
                        
                        with open('C:/Users/user/rep.txt', 'a') as file:
                            file.write(str(headline.group(1)) + '\n' + str(authors) + '\n' + 
                                    str(list_of_new_url[i]) + '\n\n')
                        print(str(headline.group(1)) + '\n' + str(authors) + '\n' + 
                            str(list_of_new_url[i]) + '\n\n')
                        
                        break 
                        
                        
    if len(list_of_new_url) != 0:
        old_url = list_of_new_url[0]
    print('-------------------------------------------------------------------------------')
                        
    # задержка на час
    time.sleep(3600) 
        
      
    
```
